```python
# import required libraries
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
# load data
dataset_0 = pd.read_csv('grades_dataset.csv')
print(dataset_0)
```

```python
# load data
dataset = pd.read_csv('grades_clustering.csv')
print(dataset)
```

```python
# create kmeans object and call fit
KM = KMeans(n_clusters=2, init='k-means++', random_state=170)
KM = KM.fit(dataset)
print("The cluster centroids are: \n", KM.cluster_centers_)
print("Cluster_label:\n", KM.labels_)
print("Sum of distances of samples to their closest cluster center: \n", KM.inertia_)
```

```python
# visualize datapoints' assigned cluster in 2-D
colors = ['blue','red']
plt.scatter(dataset.grade, 
            dataset.attend_class,
            c=KM.labels_, 
            cmap=matplotlib.colors.ListedColormap(colors), 
            s=75)
plt.show()
```

```python
# we can also choose another two features to visualize
plt.scatter(dataset.extra_hours, 
            dataset.attend_class, 
            c=KM.labels_, 
            cmap=matplotlib.colors.ListedColormap(colors), 
            s=75)
plt.show()
```

```python
# we can also choose another two features to visualize
plt.scatter(dataset.grade, 
            dataset.extra_hours, 
            c=KM.labels_, 
            cmap=matplotlib.colors.ListedColormap(colors), 
            s=75)
plt.show()
```

```python
# or we can visualize in 3-D
fig = plt.figure(1, figsize=(5, 5))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(dataset.grade, 
           dataset.attend_class, 
           dataset.extra_hours, 
           c=KM.labels_, 
           cmap=matplotlib.colors.ListedColormap(colors),
           s=75)
plt.show()
```

```python
# tuning n_clusters
KM_2 = KMeans(n_clusters=3, init='k-means++', random_state=170)
KM_2 = KM_2.fit(dataset)
print("The cluster centroids are: \n", KM_2.cluster_centers_)
print("Cluster_label:\n", KM_2.labels_)
print("Sum of distances of samples to their closest cluster center: \n", KM_2.inertia_)
```

```python
# AND visualize in 3-D
colors_3 = ['blue','red','yellow']
fig = plt.figure(1, figsize=(5, 5))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(dataset.grade, 
           dataset.attend_class, 
           dataset.extra_hours, 
           c=KM_2.labels_, 
           cmap=matplotlib.colors.ListedColormap(colors_3),
           s=75)
plt.show()
```
